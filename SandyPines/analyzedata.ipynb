```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('sandypines.csv')
df.sample(5)
```

```python
df_sorted = df.sort_values(by='Price', ascending=False)
```

```python
x = df_sorted['Year']
y = df_sorted['Price']
plt.scatter(x, y)
plt.xlabel('Year')
plt.ylabel('Price')
plt.show()
```

```python
sns.heatmap(data=df_sorted)
```
